```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("homework.csv")
df.head()
```

```python
X = df[['lotsize']]          # Variable indépendante
Y = df['price']              # Variable dépendante

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
model_simple = LinearRegression()
model_simple.fit(X_train, Y_train)
```

```python
Y_pred = model_simple.predict(X_test)

print("MSE :", mean_squared_error(Y_test, Y_pred))
print("R2 :", r2_score(Y_test, Y_pred))
```

```python
plt.scatter(df['lotsize'], df['price'], color='blue')
plt.plot(df['lotsize'], model_simple.predict(df[['lotsize']]), color='red')
plt.xlabel("Lot Size")
plt.ylabel("Price")
plt.title("Simple Linear Regression (lotsize → price)")
plt.show()
```

```python
# Transformer yes/no → 1/0 (méthode moderne sans warning)
df_encoded = df.copy()

binary_cols = ["driveway", "recroom", "fullbase", "gashw", "airco", "prefarea"]

for col in binary_cols:
    df_encoded[col] = df_encoded[col].map({"yes": 1, "no": 0})
```

```python
X = df_encoded.drop(columns=['price'])
Y = df_encoded['price']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
model_multi = LinearRegression()
model_multi.fit(X_train, Y_train)
```

```python
coefficients = pd.DataFrame({
    'Variable': X.columns,
    'Coefficient': model_multi.coef_
})

print("Intercept :", model_multi.intercept_)
coefficients
```

```python
Y_pred_multi = model_multi.predict(X_test)

print("MSE :", mean_squared_error(Y_test, Y_pred_multi))
print("R2 :", r2_score(Y_test, Y_pred_multi))
```
